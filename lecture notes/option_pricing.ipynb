### Introduction

The terms **derivatives pricing** and **option pricing** refer to techniques for pricing assets connected by **arbitrage** relationships.  An **arbitrage opportunity** occurs when a trading strategy generates gains with positive probability (in the present or the future) but generates losses with zero probability. 

In practice, many so-called arbitrage relationships are only approximate because the assets' risk do not fully offset one another, leaving a small amount of risk behind.  Trading strategies which try to exploit arbitrage opportunities typically involve offsetting long and short positions in which the short positions **hedge** the risks of the long positions, and vice versa.

Examples of using arbitrage relationships to price assets include pricing: 

1. **Futures contracts** and **forward contracts**, including **foreign exchange**. 

2. **Put options** and **call options**.

3. Risky debt governed by probabilities of default.

4. Riskfree (e.g., government) bonds governed by movements in the term structure of interest rates.

5. **Real options**, such as oil wells closely related to the price of oil.

6. **Executive compensation**, such as CEO compensation tied to the price of a company's stock.

7. **Warrants** (long-term call options on stock which dilute outstanding shares) and **convertible bonds** (debt convertible in a firm's equity).



### Forward contracts for currencies

A **forward contract** is a contractual agreement to exchange financial assets at a future date at a fixed price determined when the contract is negotiated.

It is best illustrated with a simple example: Consider a contract in which I agree to pay $P_f$ (USD / EUR) for Euros $m$ (years) from now. Suppose the current exchange rate is $P$ (USD/EUR), the current U.S. dollar interest rate is $r_{\text{US}}$ (annual, continuously compounded) and the current Euro interest rate is $r_{\text{EUR}}$ (annual, continuously compounded).

To think through the arbitrage relationship, consider two ways of investing U.S. dollars today to obtain Euros in the future:

1. Invest dollars to buy dollar-denominated fixed income securities now, then convert to Euros at the forward rate $P_f$ (USD/EUR) negotiated today.  This yields $\text{e}^{r_{\text{USD}} \sm m} / P_f$ (EUR/USD at maturity).

2. Buy Euros today, obtaining $1 / P$ (EUR per USD) today at the spot rate, then invest the Euros in EUR fixed income securities.  This yields $\frac{1}{P} \sm \text{e}^{r_{\text{EUR}} \sm m}$ (EUR/USD in the future).

The difference between these two quantities can be represented in at least four different ways:

*Euros in the future*: The difference between the two previous quantities is the net future value $NFV_{\text{EUR/USD}}$ (EUR/USD at maturity) of a forward contract with forward exchange rate $P_f$:

$$
NFV_{\text{EUR/USD}} = \text{e}^{r_{\text{USD}} \sm m} \sm \frac{1}{P_f} - \frac{1}{P} \sm \text{e}^{r_{\text{EUR}} \sm m} .
$$

*Euros today*: This expression can be converted into a present value $NPV_{\text{EUR/USD}}$ by multiplying by the present value of a Euro $\text{e}^{-r_{\text{EUR}} \sm m}$ to obtain

$$
NPV_{\text{EUR/USD}} = \frac{1}{P_f} \sm \text{e}^{(r_{\text{USD}} -r_{\text{EUR}}) \sm m} - \frac{1}{P}.
$$

*Dollars today*: The net present value in dollars today (per dollar invested) is obtained by multiplying both sides of the previous equation by the exchange rate $P$:

$$
NPV_{\text{USD/USD}} = \frac{P}{P_f} \sm \text{e}^{(r_{\text{USD}} - r_{\text{EUR}}) \sm m} - 1.
$$

*Dollars in the future*: By dividing by the future value of a dollar $\text{e}^{r_{\text{USD}} \sm m}$, this expression is further converted into the future value in dollars as a fraction of dollars invested today:

$$
NFV_{\text{USD/USD}} = \frac{P}{P_f} \sm \text{e}^{-r_{\text{EUR}} \sm m} -  \text{e}^{-r_{\text{USD}} \sm m}.
$$

Obviously, if one of the previous expressions is positive, all four of them are also posistive. Similarly, if one is negative, all are negative.

If the NPVs and NFVs are positive, I obtain more Euros by entering the forward contract and investing dollars than by buying EUR on the **spot market** and investing Euros.

The positive value represents an **arbitrage opportunity**, defined as a trading strategy which generates positive cash flow with no risk.  The four different formulas depend on whether the positive cash flow represents cash in dollars or Euros, now or in the future.

To take advantage of the arbitrage opportunity, a trader will enter into a forward contract to buy Euros at the future date, then **hedge** the forward contract by borrowing Euros at the Euro rate, using the borrowed Euros to buy dollars, and investing the dollars at the dollar interest rate. 

If all of the expressions are negative, a trading will enter into the arbitrage strategy in reverse by selling forward Euros, then hedging by borrowing dollars, using the proceeds to buy Euros, and investing the Euros at the Euro interest rate. Borrowing dollars, using the proceeds to buy Euros, and investing the Euros at the Euro interest rate **replicates** the cash flows of the forward contract.

Arbitrage trading will tend to prevent any of the four expressions from being positive or negative by more that an amount representing the costs of carrying out the arbitrage trading strategy.

The **no-arbitrage** value of the forward contract is the value of $P_f$ (EUR/USD) which makes the four equivalent expressions all zero:

$$
P_f = P \sm \text{e}^{(r_{\text{USD}} - r_{\text{EUR}}) \sm m} .
$$

This relationship is often called **(covered) interest rate parity**. The **implied forward price** $P_f$ in the above equation is the **zero NPV forward price** in the sense that a trader breaks even trading at this price.

In practice, the no-arbitrage relationship does not hold exactly because the bid price is lower than the spot price in the foreign exchange market, and the interest rate for borrowing is higher than the interest rate for lending.

The forward rate on Euros is higher than the spot rate when Euro interest rates are higher than dollar interest rates, and vice versa.


### Warning!

Although simple, interest rate parity equations are error-prone: 

1.  The same arbitrage relationship can be viewed from the perspective of investors in the two different currencies.  This **symmetry** makes it easy to express relationships backwards.  changing the **numeraire** from dollars to Euros changes the numerical description of the same relationship from EUR/USD to USD/EUR.  When the numeraire changes, the arbitrage relationship itself should not change.

2. Foreign exchange rates are **relative prices**. The same relative price can also be expressed by its reciprocal; it is easy to confuse prices representing dollars per Euro with price representing Euros per dollar.  The USD market typically uses inconsistent ratios for different currencies.  For example, when trading British pounds (GBP) or Euros (EUR), U.S. traders typically use USD/GPB and USD/EUR, but when trading Japanese yen (JPY), U.S. traders use the reciprocal JPY/USD. Note that the USD moves from the numerator to the denominator, perhaps to make the "typical" numerical value of an exchange rate greater than one.

3. It is easy to confuse the present with the future, therefore confusing present value with future value.

4. The correct interest rate depends on the maturity (horizon) of the arbitrage trade. It is necessary to use the correct maturity from the term structure of interest rates.

5.  It is important to use an instrument appropriate for the instruments being traded. For example, a commercial bank conducting interest rate arbitrage might use rates on bank CDs to determine arbitrage relationships. In the past, banks used LIBOR rates, which were based on bank CD rates. A central bank might use government security rates rather than bank CD rates because central banks finance their activities with government securities.

6. Sometimes currencies can be confused. A "dollar" can refer to U.S. dollars (USD), Canadian dollars (CAD), or Australian dollars (AUS). Some countries peg their currencies to other currencies, in which case it is easy to use the wrong country's interest rate to price arbitrage relationships.

### What makes static arbitrage relationships fail?

Covered interest rate parity is an example of a **static** arbitrage relationship. This means all of the "trading"---arranging forward contract prices, buying or selling foreign exchange, investing or borrowing in either currency---takes place when the arbitrage trade is put into place. In theory, if an arbitrge profit is "locked in", the same profit will be harvested at the maturity of the trade regardless of how exchange rates and interest rates move.

Nevertheless, there are events which make arbitrage trades either impossible or unexpectedly unprofitable. These include:

1. Regulations, such as bank **capital requirements** or **leverage requirements**, which affect the ability of arbitragers to carry out the necessary trades.

2. Prohibitions on trades, including **capital controls** (which prevent foreign exchange transactions) and **interest rate controls**, often directed at specific types of traders.

3. **Bank failures** or **sovereign defaults**, in which banks or governments do not pay back debts they owe.

4. Threats of **punitive taxes** or **expropriation**, such as the U.S. government expropriating Russian central bank and Russian oligarchs' assets and the Russian government expropriating foreign assets in the aftermath of Russia's invasion of Ukraine.

5. **Counterparty defaults**, in which one party to a transaction does not honor its contractual commitments, such as settling an under-water forward exchange transaction as contractually required.

When capital controls are in place, there are often multiple foreign exchange rates within a country, with internal exchange rates deviating from open market (external) exchange rates. For example, favored entities may have access to buying foreign currency at more favorable rates than other entities. While such policies may be powerful tools for managing the macroeconomy, they are obviously prone to corruption and cronyism.

### Market probabilities

A bet on an election or a sporting event can be interpreted as a security with a payoff of one dollar if the favored candidate or home team wins, zero dollars if the home team loses. Such bets with zero-one payoffs are called **pure securities**.  If the odds are two-to-one in favor of the home team winning, the price of the security is approximately 2/3 of a dollar. This approximation ignores the time value of money by treating the bet as a forward contract to be settled when the outcome is determined.

While the market price of 2/3 of a dollar can be interpreted as a probability, it is not necessarily the **actuarily fair** (**objectively correct** or **true**) probability. For example, the actuarily fair probability of the home team winning might be 1/2, which implies a price of 50 cents (1/2 of a dollar) for the pure security. The market-implied probability of about 66.667 cents might exceed the actuarily fair probability of 50 cents because the home team supporters are exceptionally and irrationally enthusiastic about the home tema winning, compared with the other team's supporters. Researchers believe that larger and richer sports markets (say New York city versus Cincinatti, OH) tend to push market odds against the large market winning. There is also a debate about whether **prediction market** prices for election outcomes are biased in favor of Republicans winning, presumably because Republicans either enjoy participating in markets more than Democrats or because Republicans have more money to bet.

Finance researchers use the term **risk neutral probabilities** to describe market-implied probabilities. The intuition for the terminology is that the market-implied probabilities would equal the actuarily fair probabilities if market participants were all risk neutral (risk aversion parameter $\rho \rightarrow 0$) and had infinitely accurate actuarial skills. The terminology **risk neutral probability** is confusing: Actual market-implied probabilities are not generally the actuarily fair probabilities because the market is not really risk neutral. 


### Observability

The is a hugely important difference between market probabilities and actuarily fair probabilities:  

**Market probabilities can be directly observed or inferred from market prices, but actuarily fair probabilities cannot be inferred from prices.  In this sense, they are not observable and must be estimated from models with data.**

There are other important finance examples of the general principle that **market-implied** predictions can be observed but **objectively correct** predictions cannot be observed easily. These finance examples all require approximate models.

1. **Yield to maturity**: If a **defaultable** corporate bond has a **promised** yield to maturity of $r = 6.00$ percent per year while a default-free government security has a promised yield to maturity of $r_f = 4.00$ percent per year, the market-implied losses from default are expected to be approximately $r - r_f = 2.00$ percent per year. The actuarily fair losses might be different; finance research typically suggests that market-implied expected losses are higher than actuarily fair losses. A trader who buys defaultable debt and hedges the interest rate exposure by selling or shorting risk-free debt is betting that risky bonds will not default.

2. **Gordon's growth formula**; If the risk-free rate $r_f$, current dividend $D$, price of a security $P$  security beta $\beta$, and risk premium on the market $\pi$ are all observed, Gordon's growth formula $P = D / (r_f + \beta \sm \pi - g)$, can be inverted to obtain an **implied growth rate** 
$$
g = r_f + \beta \sm \pi - \frac{D}{P} .
$$
In practice, the risk premium on the market portfolio, like actuarily fair probabilities, $\pi$ is very difficult to observe. This fact makes the inverted formula difficult to apply in practice and explains why there is a great deal of interest in estimating the risk premium on the market portfolio $\pi$.

3. **Black-Scholes model of option pricing**: Put and call options have higher prices when market participants expect the underlying assets to have higher volatility.  Given a model of how options are priced, we can use the model to derive an **implied volatility** from market prices of options. We will use the **Black-Scholes model** of option pricing to do this.



### Put Options and Call Options

A **call option** is a contract in which the buyer of the call has the right, but not the obligation, to purchase an asset at a contractually fixed **strike price** on (or perhaps before) a contractually fixed future date.

A **put option** is a contract in which the buyer of the call has the right, but not the obligation, to sell an asset at a contractually fixed **strike price** on (or perhaps before) a contractually fixed future date.

The seller of an option is said to **write** the option. If the buyer or a call exercises the right to buy the asset or the buyer of a put option exercises the right to sell the asset, the option buyer is said to **exercise** the option. If the option contract allows for **early exercise** (exercise before the expiration date of the option), historical practice says the contract is an **American** option. If early exercise is not allowed, the option is a **European** option.

Unlike stocks or commodities, options are in **zero net supply**. This means that for number of contracts bought is always equal to the number of contracts sold.  Thus, when options is exercised, there must exist option writers who are contractually obligation to sell the asset at the strike price when a call is exercised or to buy the asset at the strike price when a put is exercised.





### Payoffs of puts and calls

We will use the following notation:

$$P(t) = \text{price of risky asset at date } t,$$

$$K = \text{strike price of option},$$

$$T = \text{expiration date of opt,ion}.$$

Here, $P(t)$ changes randomly over time; $K$ and $T$ are contractually fixed quantities which define the terms of the option contract. 

At the expiration date, the **exercise value** of a call option is $\max[P(T) - K, 0]$. Why? 

* If $P(T) > K$ at expiriation, the buyer of the call option can buy the asset at price $K$ by exercising the call, then turn around and sell the asset for $P(T)$, earning a profit of $P(T) - K$. The buyer of the call option will profit from exercising in this way, even if he does not want to own the asset at all. 

* If $P(T) < K$ at expiration, the buyer of the call will let the option expire worthless; if he really wants to buy the asset, it is cheaper to pay $P(t)$ in the market than to exercise and pay $K$.

Similarly, at the expiration date, the **exercise value** of a put option is $\max[K - P(T), 0]$. Why? 

* If $P(T) < K$ at expiriation, the buyer of the put option can sell the asset at price $K$ by exercising the put, then turn around and buy the asset for $P(T)$, earning a profit of $K - P(T)$. The buyer of the put option will profit from exercising in this way, even if he owns the asset and does not want to sell it. 

* If $P(T) > K$ at expiration, the buyer of the put will let the option expire worthless; if he really wants to sell the asset, it is better to sell at the higher price $P(t)$ in the market than to exercise and sell at the lower price $K$.

For fixed strike price $K$, These exercise values can be thought of a functions of the price of the asset at the exercise date $T$.

It is easy to plot these values:


```python
import numpy as np
import matplotlib.pyplot as plt

K = 100.00  # dollars per share
T = 0.25  # 3 months
Pmin = 50.00
Pmax = 150.00
Pt = np.linspace(start=Pmin, stop=Pmax, num=101, endpoint=True)
long_call_payoff = np.maximum(Pt - K, 
                              0.00)
long_put_payoff = np.maximum(K - Pt, 0.00)
short_call_payoff = -np.maximum(Pt - K, 0.00)
short_put_payoff = -np.maximum(K - Pt, 0.00)

fig, ax = plt.subplots(2, 2, figsize=(6, 6))

linewidth = 3.00
markersize = 1.00

ax[0,0].plot(Pt, long_call_payoff, 'b-', markersize=markersize, linewidth=linewidth)
ax[0,0].set_title("Long Call Payoff\n monotonically increasing \n convex")
# ax[0].set_xlabel("$P(T) =$ asset price at expiration" )
ax[0,0].set_xlabel("$P(T)$" )
# ax[0].set_ylabel("$\max[P(T) - K, 0]$ = call exercise value")
ax[0,0].set_ylabel("$\max[P(T) - K, 0]$")

ax[0,1].plot(Pt, long_put_payoff, 'b-', markersize=markersize, linewidth=linewidth)
ax[0,1].set_title("Long Put Payoff\n monotonically decreasing \n convex")
# ax[1].set_xlabel("$P(T) = asset price at expiration")
ax[0,1].set_xlabel("$P(T)$")
# ax[1].set_ylabel("$\max[K - P(T), 0]$ = put exercise value")
ax[0,1].set_ylabel("$\max[K - P(T), 0]$")

ax[1,0].plot(Pt, short_call_payoff, 'b-', markersize=markersize, linewidth=linewidth)
ax[1,0].set_title("Short Call Payoff\n monotonically decreasing \n concave")
# ax[0].set_xlabel("$P(T) =$ asset price at expiration" )
ax[1,0].set_xlabel("$P(T)$" )
# ax[0].set_ylabel("$\max[P(T) - K, 0]$ = call exercise value")
ax[1,0].set_ylabel("$-\max[P(T) - K, 0]$")

ax[1,1].plot(Pt, short_put_payoff, 'b-', markersize=markersize, linewidth=linewidth)
ax[1,1].set_title("Short Put Payoff\n monotonically increasing \n concave")
# ax[1].set_xlabel("$P(T) = asset price at expiration")
ax[1,1].set_xlabel("$P(T)$")
# ax[1].set_ylabel("$\max[K - P(T), 0]$ = put exercise value")
ax[1,1].set_ylabel("$-\max[K - P(T), 0]$")

plt.tight_layout() # Prevents plots from being squashed together too much.

plt.show()
```

### Monotonicity and convexity

The four plots above show that the payoff on long positions in both puts and calls is nonnegative. The call payoff is strictly positive for high enough asset price. The put payoff is strictly positive for low enough asset price (assuming $K > 0$).  This implies that the market price of a call or put will generally be positive.  The positive price paid for the option is omitted from the graphs. In practice, the option is likely to have traded at different prices at different times, even though the contractual terms of the option contract $K$ and $T$ remain fixed. 

The graph of the long call option payoff is monotonically increasing (weakly), and the graph of the long put option payoff is monotonically decreasing (weakly). In other words, a long call position is a **bullish** bet on the value of the underlying asset, and a long put position is a **bearish** bet on the value of the underlying asset.  For the seller of the option, this is obviously reversed because the any profit a buyer earns upon exercise represents and equal and opposite loss to the seller of the option.  Thus, a short call position is a **bearish** bet on the value of the asset, and a short put position is a **bullish** bet on the value of the asset.

Both long call and long put payoffs are **convex** functions of the value of the asset at expiration. Similarly, both short call and short put positions are **concave** function of the asset value at expiration. 

The convexity of the long call and put option payoffs and the concavity of short call and put payoffs can be seen from the definitions of **convexity** and **concavity**.  

* A function $f(x)$ is **convex** if all points on a line segment connecting two points on the graph of the function lie on or **above** the graph of the function. 

* A function $f(x)$ is **concave** if the function $-f(x)$ is convex. Alternatively, a function is concave if all points on a line segment connecting two points on the graph of the function lie on or **below** the graph of the function.

To summarize:

* With respect to bullishness and bearishness, calls and puts have opposite properties because long call positions represent bullish bets while long put positions represent bearish bets.

* With respect to convexity and concavity, calls and puts have similar properties because long positions in both calls and puts have convex payoffs.




```python
import numpy as np
import matplotlib.pyplot as plt

K = 100.00  # dollars per share
T = 0.25  # 3 months
Pmin = 50.00
Pmax = 150.00
Pt = np.linspace(start=Pmin, stop=Pmax, num=101, endpoint=True)
long_call_payoff = np.maximum(Pt - K, 0.00)
long_put_payoff = np.maximum(K - Pt, 0.00)
short_call_payoff = -np.maximum(Pt - K, 0.00)
short_put_payoff = -np.maximum(K - Pt, 0.00)

p0 = 75.00
p1 = 125.00

c0 = np.maximum(p0 - K, 0.00)
c1 = np.maximum(p1 - K, 0.00)

alpha = np.linspace(start=0.00, stop=1.00, num=101, endpoint=True)
line_segment_pvals = (1.00 - alpha) * p0 + alpha * p1
line_segment_cvals = (1.00 - alpha) * c0 + alpha * c1

fig, ax = plt.subplots(1, 1, figsize=(5, 5))

linewidth = 3.00
markersize = 1.00

ax.plot(Pt, long_call_payoff, 'b-', markersize=markersize, linewidth=linewidth)
ax.plot(line_segment_pvals, line_segment_cvals, 'r-',  linewidth=1.00)
ax.plot((p0, p1), (c0, c1), 'ro',  markersize=5.00)
ax.set_title("Long Call Payoff\n monotonically increasing \n convex")
ax.set_xlabel("$P(T)$" )
ax.set_ylabel("$\max[P(T) - K, 0]$")

plt.tight_layout() # Prevents plots from being squashed together too much.

plt.show()
```

### Interpretation as Jensen's Inequality

The fact that the red line segment in the plot above lies above the graph of the payoff on the call option has a probabilistic interpretation related to Jensen's inequality.  

Suppose that the stock price has only two outcomes `p0=75.00` with probability `1.00-alpha` and `p1=125.00` with probability `alpha`.

* The red line segment represents the expected payoff on the call option for different values of the probability $\alpha$ in the interval $[0,1]$. 

* The exercise value below any point on the red line represents the payoff associated with the expected price `(1.00-alpha)*p0 + alpha*p1`.

Jensen's inequality says that since the payoff of the call option is a convex function, the red line segment lies above or on the graph of the payoff function.  This interpretation of the plot shows that **Jensen's inequality is closely related to the definition of convexity**.



### Volatility

The graph also has an interpretation related to the volatility of the underlying asset price.

If we move the two points `p0` and `p1` further apart by moving `p0` to the left (say from `p0=75.00` to `p0=60.00`) and moving `p1` to the right (say from `p1=125.00` to `p1=140.00`), then the red line segment moves up and gets further away from the graph of the payoff on the call option. 

Moving `p0` and `p1` further apart can be intuitively interpreted as increasing the volatility of the asset's payoff. This example therefore appears to illustrate the the more general principle that increasing the volatility of the assets payoff increases the expected payoff on an option. This intuition, while reasonable, is not mathematically precisely correct. 


### Increasing risk (optional, advanced)

What is the correct mathematical way to define "increasing risk" to guarantee that the expected payoff on calls and puts increases?

The correct mathematical intuition is that randomizing `p0` and `p1` in a manner which preserves their respective means, which is called a **mean-preserving spread**, increases the expected payoff on the option. In the context of the example above, this means replacing `p0=75.00` with a random variable which attains new values of `60.00` and `140.00` with probabilities precisely chosen to make the expected value equal to 75.00. Similarly, `p1=125.00` is replaced with a random variable which attains the new values of `60.00` and `140.00` with probabilities chosen to make the expected value remain equal to 125.00.  Randomizing in this manner is guaranteed to increase the expected payoff on the option, as the example illustrates.  

To understand this point further, let $x$ and $\epsilon$ be random variables, and assume $\E[\epsilon]=0$. Now define the random variable $y$ by $y = x + \epsilon$.  If we think of $\epsilon$ as "noise" added to $y$, then intuition suggests that, under some reasonable assumptions, the expected payoff on a call or put option will increase if the distribution of the payoff is changed from that of $x$ to that of $y$ by adding noise $\epsilon$. 

What precise assumptions on the joint distribution of $x$ and $\epsilon$ are needed to ensure that that the expected payoff increases if the distribution is changed from that of $x$ to that of $y$? Here are three possibilities:

1. Suppose that $x$ and $\epsilon$ are **independently distributed**. Then indeed, if $f$ is any convex function, such as the payoff on a call or put, then we must have $\E[f(y)] \ge \E[f(x)]$. This assumption gives us our result but is stronger than what we need.

2. Suppose that $x$ and $\epsilon$ are **uncorrelated**. Then it is mathematically possible to concoct examples where $\E[f(y)] \ge \E[f(x)]$ is false. This point is the same as the point that increasing variable does not guarantee $\E[f(y)] \ge \E[f(x)]$. This assumption does not give us the result we want; it is too weak.

3. Suppose that $\E[\epsilon \vert x] = 0$ for all realizations of $x$. Then $\E[y \vert x] = x$ and $\var[y] = \var[x] + \var[\epsilon]$. This assumption guarantees the desired result $\E[f(y)] \ge \E[f(x)]$. Furthermore, it can be shown mathematically that this is exactly the assumption that we need to obtain the result. In other words, the correct way to define "adding noise to $x$" is to replace $x$ with $y := x + \epsilon$, where $\epsilon$ follows the **fair game** assumption $\E[\epsilon \vert x] = 0$. A **fair game** is a gamble whose expected payoff is zero. This implies cumulative net winnings follow a **martingale**. Mathematically, adding $\epsilon$ in this way is conceptually the same as the mean-preserving spread. It is guaranteed to make the expected payoff on a call option or put option increase (weakly). Furthermore, it can be shown that any change in distribution which increases (weakly) the expected payoff on every convex function must satisfy the fair game assumption. 

To summarize, the fair game assumption is weaker than the assumption that $x$ and $\epsilon$ are independently distribution. It is stronger than the assumption that $x$ and $\epsilon$ are uncorrelated.  The fair game assumption is exactly what is needed to guarantee that increasing the risk of the distribution (from $x$ to $y=x + \epsilon$) increases the expected payoff on call and put options (or any other convex payoff function) when "noise" is added to $x$.

 

It is theoretically possible to concoct mathematical counterexamples in which increasing the variance of a probability distribution of underlying asset prices decreases the expected payoff on the option. These counterexamples require probability distributions different from normal or lognormal. 





### The problem of pricing options

Consider the problem of valuing a put or call option. Intuitively, the convexity of the option implies that the price should be higher if the volatility is higher. To derive a model, we must therefore make some assumptions about volatility. 

The obvious assumption to make is one we have made before: **The stock's price follows geometric brownian motion**. Mathematically, we can write the geometric brownian motion assumption as

$$
P(t) = P_0 \sm \text{e}^{(r_f + \beta \sm \pi - \sigma^2 / 2) t + \sigma \sm B(t)}.
$$

A call option with strike price $K$ and maturity $m = T - t$ has a payoff at date $T = t + m$ given by

$$
C(T) = \max [P(T) - K, 0].
$$

Here we assume that **early exercise** of the option is not allowed. It can only be exercised at maturity (**European option**), not early (**American option**).  We also assume that the stock does not pay dividends.

Before the Black-Scholes model became known in the 1970s, finance researchers tried unsuccessfully to derive the value of a call option before expiration. One somewhat misguided way to think about the problem is to use CAPM intuition to write the option price as the discounted value of the future cash flow, using a discount rate which takes into account the CAPM beta of the option $\beta_c$:

$$
C(T - m) = \mathrm{E} \left[ \mathrm{e}^{-(r_f + \beta_c \sm \pi) \sm m} \max [P(T) - K, 0] \; \vert \; P(t) \right].
$$

This approach is misguided because the unknown beta of the option $\beta_c$ is not a constant; it changes randomly over time, and there is no obvious way to derive a value for $\beta_c$. 

The option has embedded leverage; this makes the beta greater than one. If the call option is **deep out-of-the-money** (strike price far above market price for stock), the options embeds a great deal of leverage, the beta is very large, and the option price is very sensitive to the stock price. If the call option is **deep in-the-money** (strike price far below market price for stock), the option does not embed much leverage, the beta is close to the stock's beta (ignoring interest rate effects), and the option price moves approximately one-for-one with the stock price. Since the stock price changes randomly over its life, the option's beta also changes randomly, even if the stock's beta is itself constant.  This makes it difficult to determine the constant value of $\beta_c$ which would make the above present-value formula correct.  Before the 1970s, finance researcher got stuck here. Determining the correct value for $\beta_c$ is just as difficult a problem as valuing the option itself.








### Black-Scholes formula

An obvious way to avoid having to calculate an "average beta" $\beta_c$ for the option is to assume $\pi=0$, which says that the risk premium on both the stock and the market portfolio is zero. The geometric brownian motion simplifies to

$$
P(t) = P_0 \sm \text{e}^{(r_f + \sigma^2 / 2) t + \sigma \sm B(t)}.
$$

Now the $\beta_c$ term disappears from the call option formula. We can price the option in a **risk neutral** manner by discounting its final cash flow at the riskfree rate. We obtain, for $m = T - t$,

$$
\begin{aligned}
C(t) &= \mathrm{E} \left[ \mathrm{e}^{-r_f \sm m} \sm \max [P(T) - K, 0] \; \vert \; P(t) \right]\\
&=
\mathrm{E} \left[ \mathrm{e}^{-r_f \sm m} \sm \max [P(0) \sm \text{e}^{(r_f - \sigma^2 / 2) \sm m + \sigma \sm B(T)} - K, 0] \; \vert \; P(t) \right] \\
&=
\mathrm{E} \left[ \mathrm{e}^{-r_f \sm m} \sm \max [P(t) \sm \text{e}^{(r_f - \sigma^2 / 2) \sm m + \sigma \sm (B(T) - B(t))} - K, 0] \right]. \\
&=
\mathrm{E} \left[ \max [P(t) \sm \text{e}^{-\sigma^2 \sm  m / 2 + \sigma \sm (B(T) - B(t))} -   K \sm \mathrm{e}^{-r_f \sm m}, 0] \right].
\end{aligned}
$$

In the above equations, the second equation is obtained by substituting in the geometric brownian motion with $\pi=0$, the third equation incorporates the effect of conditioning on the known price $P(t)$, and the last equation simplifies calculations with the risk-free rate.

The equation can be expressed as an integration problem by observing that $(B(T) - B(t)) / \sqrt{m} \sim \mathcal{N}(0, 1)$. Since the strike price $K$ satisfies 

$$
\begin{aligned}
K &= P(t) \sm \text{e}^{(r_f + \sigma^2 / 2) \sm m + \sigma \sm (B(T) - B(t))}\\
&= P(t) \sm \text{e}^{(r_f + \sigma^2 / 2) \sm m + \sigma \sm \sqrt{m} \sm z},
\end{aligned}
$$

the value of $z$ corresponding to the strike price, denoted $z_k$, is

$$
z_k = \frac{\log \left( \frac{K}{P} \right) - (r_f + \sigma^2 / 2) \sm m}{\sigma \sm \sqrt{m}}.
$$


We can write the value as a function of five parameters $P=P(t)$, $K$, $m$, $r_f$, and $\sigma$. The result is the integration problem

$$
\begin{aligned}
C(P, K, m, r_f, \sigma)  
&= \int_{z=-\infty}^{+\infty} \: 
\max [P \sm \text{e}^{-\sigma^2 m / 2 + \sigma \sm \sqrt{m} \sm z} -   K \sm \mathrm{e}^{-r_f \sm m}, 0]
\sm \frac{1}{\sqrt{2 \pi}} \sm \mathrm{e}^{-z^2 / 2} \mathrm{d}z.  \\
&= \int_{z=z_c}^{+\infty} \: 
\left( P \sm \text{e}^{-\sigma^2 m / 2 + \sigma \sm \sqrt{m} \sm z} -   K  \sm \mathrm{e}^{-r_f \sm m} \right)
\sm \frac{1}{\sqrt{2 \pi}} \sm \mathrm{e}^{-z^2 / 2} \mathrm{d}z. 
\end{aligned}
$$

The integration problem has a closed-form solution, in terms of the cdf of a standard normal distribution, $\mathrm{N}(z) = \int_{-\infty}^{z} \tfrac{1}{\sqrt{2 \sm \pi}} \sm \mathrm{e}^{-z^2/2} \mathrm{d}z$, obtained by solving a tedious integral problem at the level of BC calculus.  The solution is called the **Black-Scholes formula**:

$$
\begin{aligned}
C(P, K, m, r_f, \sigma) &= P \sm \mathrm{N}(d_1) - K \sm \mathrm{e}^{-r_f \sm m} \sm \mathrm{N}(d_2), \\
\text{where} \qquad d_1 &:= \frac{\log \left( \frac{P}{K \sm \mathrm{e}^{-r_f \sm m}} \right) + \sigma^2 \sm m / 2}{\sigma \sm \sqrt{m}},
\qquad
d_2 := d_1 - \sigma \sm\sqrt{m}.
\end{aligned}
$$

Before the **Black-Scholes model** was discovered in the 1970s, finance practitioners knew about this formula and used it. Both practitioners and researchers thought that it was not the theoretically correct solution for the price of an option but instead only a useful approximation.



### Black-Scholes model

The contribution of Black, Scholes, and Merton was to show that what we now call the Black-Scholes formula is indeed the correct solution for the **no-arbitrage** price of a call option on a stock whose price follows geometric Brownian motion. The **no-arbitrage price** is the only price for the option which does not allow a dynamic trading strategy which creates an arbitrage profit.

Black and Scholes used the following intuition to derive the Black-Scholes model: A trader can buy or sell an option, then hedge the exposure to the stock's price by selling or buying exactly enough of the underlying stock to the hedge the option's exposure to the risk resulting from fluctuations in the stock price. Since the beta of the option changes randomly over time, the correct hedge ratio (called the option's **delta**) also changes over time. This requires that the trader dynamically update the hedge over time. Of course, infrequent updating makes the hedge imperfect (i.e., it introduces risk) because the option price is a **nonlinear** function of the stock's price. Black and Scholes observed that the hedge becomes perfect in the limit as the hedge is updated continuously. 

In a **perfect hedge** (my terminology), the portfolio starts with the option, a long or short position in the stock, and enough cash or borrowing to make the initial value zero.  The trader adjusts the cash and stock in the hedge position over time to hedge the value of the option (but does not change the option holding itself). In a perfect hedge, the value of the portfolio at maturity is nonstochastic. If we paid a price for the option that was too low, we will have a positive constant dollar amount in the portfolio at maturity. This represents an arbitrage profit.  Similarly, if we paid a price for the option that was too high, we will have a negative constant amount in the portfolio at maturity. Then we can run the hedging strategy in reverse by initially selling the option and then hedging the sale with the strategy which reverses the hedging strategy for a long position in the option.  Since this reverse strategy will yield a positive amount at maturity, it also generates an arbitrage profit.

The only initial option price which does not generate an arbitrage profit is the correct, **no-arbitrage price** we are looking for.  The intuition for the no-arbitrage price being correct is that traders would exploit arbitrage opportunities if they existed. They would buy underpriced options or sell overpriced options, hedging the risks to generate an arbitrage profit, driving the option's price towards the no-arbitrage value.

Compared with the misguided approach suggested above, the no-arbitrage approach of Black and Scholes correctly deals with the fact that the sensitivity of the option's price to the underlying stock changes randomly over time.

How do we convert the intuition of a dynamic hedging strategy into a practical algorithm for obtaining the no-arbitrage price of the option?  Black and Scholes used a different strategy from Merton to deal with this issue.

### Merton's derivation of the Black-Scholes model

Merton observed that since the hedge is perfect, it will result in the same constant portfolio value at maturity regardless of the actuarily fair value of the risk premium $\pi$.  Therefore, if we simulate the stock's price and the correct hedging strategy with an assumed risk premium of zero, we should obtain the same constant amount at maturity as we would obtain from simulating the stock with any other risk premium, including the actuarily correct one. This occurs because the option's gains from the risk premium will be exactly offset by hedge losses from the risk premium. If we pretend the risk premium is zero, then we are pretending that the market prices the asset in a risk neutral manner. This implies that we can skip the simulation and instead calculate the correct price by discounting the expected random payoff value on the option at the risk-free rate, ignoring the risk premium.  The Black-Scholes formula above already does this for us. Therefore this logic proves that the Black-Scholes formula gives us the correct, no-arbitrage value for a call option.

### Plotting the Black-Scholes formula

The following cell plots the value of a call option given by the Black-Scholes formula when the risk-free interest rate is zero.  Note the following properties:

* The call price is always greater than the exercise price. This occurs because volatility increases the value of the option.  You can verify that in the limit as $\sigma \rightarrow 0$, the Black-Scholes price converges to the exercise value. A trader would not want to exercise the call option early.

* The call price approaches zero for low stock prices.  This occurs because the probability of exercising the option goes to zero due to the low probability of huge declines in the stock price before expiration.

* The call prices approaches the exercise value $P-K$ for large stock prices. This occurs because the probability of not exercising the option goes to zero when the stock price is very large. If the risk-free rate is positive, the call price will converge to $P - K \sm \e^{-r_f \sm m}$ for large values of $P$.

* The Black-Scholes formula defines a convex function.  This is because convexity is preserved when the stock price is random in the manner defined by the Black-Scholes model.

* The function defined by the Black-Scholes formula looks "smooth".  In fact, it is infinitely differentiable.

* The slope of the function defined by the Black-Scholes formula increases from zero to one. This slope defines the sensitivity of the call option value to the stock price, which is called the option's **delta**. 

* If the maturity is decreased, the value of the option will move closer to the exercise value boundary for all underlying stock prices $P$. If the stock price does not change, the value of the option will decay over time because there is less time left before expiration for volatility to make the stock price change. 



```python
import numpy as np
import matplotlib.pyplot as plt
import math

K = 100.00  # dollars per share
m = 1.00  # one year
rf = 0.00
sigma = 0.35

def f_black_scholes(p, rf, k, m, sigma):
    rsqrt2 = 1.00 / math.sqrt(2.0)
    pvk = math.exp(-rf * m) * k
    sig = sigma * math.sqrt(m)
    d1 = (math.log(p / pvk) / sig) + 0.50 * sig
    d2 = d1 - sig
    nd1 = 0.50 * math.erfc(-d1 * rsqrt2)
    nd2 = 0.50 * math.erfc(-d2 * rsqrt2)
    res = p * nd1 - pvk * nd2
    return res

Pmin = 50.00
Pmax = 200.00
Pt = np.linspace(start=Pmin, stop=Pmax, num=101, endpoint=True)
long_call_payoff = np.maximum(Pt - K, 0.00)

# Intentionally slow and inefficient Python loop.  The homework asks you to define a more computationally efficient Black-Scholes function.
call_price = np.empty_like(Pt)
for n in range(len(call_price)):
    call_price[n] = f_black_scholes(Pt[n], rf, K, m, sigma)

fig, ax = plt.subplots(1, 1, figsize=(6, 6))

linewidth = 3.00
markersize = 1.00

ax.plot(Pt, long_call_payoff, 'b-', markersize=markersize, linewidth=linewidth)
ax.plot(Pt, call_price, 'r-',  linewidth=1.00)
ax.set_title("Black-Scholes Formula")
ax.set_xlabel("$P(T)$" )
ax.set_ylabel("Call price and expiration value")

plt.tight_layout() # Prevents plots from being squashed together too much.

plt.show()
```

### (Advanced) The Black-Scholes derivation of the Black-Scholes model

Black and Scholes used a more complicated approach based on a second-order Taylor series approximation.  Here is a slightly modified version of their approach:  

The geometric Brownian motion for $P(t)$ is obtained in the limit as $\Delta t \rightarrow 0$ in the Euler approximation 

$$
\frac{\Delta P}{P} = (r_f + \beta \sm \pi) \sm \Delta t + \sigma \sm \Delta B.
$$

The option price is some unknown nonlinear function of the stock price and time, $C(P, t)$. (The riskfree rate $r_f$, strike price $K$, and volatility $\sigma$ are not included as parameters since they are constants; only the stock price $P(t)$ and time to maturity $m = T - t$ change as time passes.) Next period's stock price can be denoted $C(P + \Delta P, t + \Delta t)$.  This stock price can be approximated to first order by using a Taylor series approximation which keeps the second-order term for $\Delta P$:

$$
C(P + \Delta P, t + \Delta t) = C(P, t) + \frac{\partial C}{\partial P} \sm \Delta P
+ \frac{\partial C}{\partial t} \sm \Delta t + \frac{1}{2} \sm \frac{\partial^2 C}{\partial P^2} \sm \Delta P^2 + o(\Delta t).
$$

Keeping the second-order term in $\Delta P^2$ is of critical importance because it captures the convexity of the option's price, which tells us how volatility affects the option's value.

Now devise a hedging strategy which at any date $t$ holds one unit (long) of the option with price $C(P_0, t_0))$, then hedges this position by selling (shorting) $h = H(P, t)$ units of the stock at price $P(t)$. The holding of the option stays constant over time but the hedge quantity $h = H(P, t)$ changes over time to reduce risk. At any given time, the positions generate cash equal to $h \sm P(t_0) - C(P_0, t_0)$. Therefore, the amount of cash flow generated by the portfolio over one time period of length $\Delta t$, including interest on the cash generated, is

$$
\Delta C - h \sm \Delta P + (h \sm P - C) \sm r_f \sm \Delta t.
$$

Substituting the Taylor series approximation (**Ito's Lemma**) for $\Delta C$ makes this

$$
\frac{\partial C}{\partial t} \sm \Delta t 
+ \frac{\partial C}{\partial P} \sm \Delta P
+ \frac{1}{2} \sm \frac{\partial^2 C}{\partial P^2} \sm \Delta P^2
- h \sm \Delta P
+ (h \sm P - C) \sm r_f \sm \Delta t
$$

To hedge risks associated with the price of the stock, set $h = \frac{\partial C}{\partial P}$. This cancels the $\Delta P$ terms, giving us

$$
\frac{\partial C}{\partial t} \sm \Delta t 
+ \frac{1}{2} \sm \frac{\partial^2 C}{\partial P^2} \sm \Delta P^2
+ \left( \frac{\partial C}{\partial P} \sm P - C \right) \sm r_f \sm \Delta t .
$$

Notice that this particular hedge ratio essentially makes the risk premium on the stock disappear because the risk premium earned on the option will be exactly offset by the risk premium lost on the short stock position.  Hedging away the risk also hedges away the risk premium.

Now $\Delta P^2 = P^2 \sm \sigma^2 \sm \Delta B^2 + o(\Delta t)$. If there are a very large number of very small intervals of length $\Delta t$, the average of many terms like $\Delta B^2$ will converge to the mean $\Delta t$. Therefore, we can replace $\Delta B^2$ with $\Delta t$, then divide by $\Delta t$ to determine the rate at which the hedge portfolio is making money at any point in time:

$$
\frac{\partial C}{\partial t} 
+ \frac{1}{2} \sm \frac{\partial^2 C}{\partial P^2} \sm P^2 \sm \sigma^2
+ \left( \frac{\partial C}{\partial P} \sm P - C \right) \sm r_f .
$$

This rate of profit is **locally nonstochastic**. If this expression is ever positive, the hedge strategy makes an arbitrage profit. If it is ever negative, running the strategy in reverse makes an arbitrage profit.  Therefore, the only arbitrage-free option pricing formula $C(P, t)$ must make this profit zero for all time periods. This give us the equation

$$
\frac{\partial C}{\partial t} 
+ \frac{1}{2} \sm \frac{\partial^2 C}{\partial P^2} \sm P^2 \sm \sigma^2
+ \left( \frac{\partial C}{\partial P} \sm P - C \right) \sm r_f 
= 0. 
$$

The above equation is a partial differential equation because it contains derivative with respect to both $P$ and $t$. It is a second-order partial differential equation because it has a second derivative with respect to $P$.

To solve the partial differential equation for the no-arbitrage price, a boundary condition is needed.  The correct boundary condition says that the price of the option at expiration date $T$ is equal to its exercise value:

$$
C(P, T) = \max[ P - K, 0].
$$

Thus far, we have reduced the problem of solving for the no-arbitrage price of a call option to solving a partial differential equation with a boundary condition.  

Black and Scholes got this far and then got stuck because they did not know how to solve the boundary value problem.  After struggling for a while, they found out that this particular partial differential equation is essentially identical to what physicists call the **heat equation**, which governs how heat dissipates in a solid.  Although solving partial differential equations is typically difficult, the solution to the heat equation is relatively simple and well known. Guess what: The solution is the same as the Black-Scholes formula!



### Replication versus hedging

The Black-Scholes arbitrage argument relies on the concept of **dynamic hedging**, which means that a long-short trader constructs a dynamic hedge which exactly offsets (hedges) the cash flos on the position to be hedged. A **dynamic hedging strategy** updates the hedge over time. This is in contrast to a **static hedging strategy**, which takes an initial position but does not update it. Covered interest arbitrage is an example of static hedging.

Another way of looking at the same idea is the concept of **dynamic replication**. Instead of buying an option, a trader constructs a dynamic trading strategy which generates the same cash flows. The dynamic replication strategy is the same as the hedging strategy multiplied by minus one: Instead of selling the stock and investing cash, the replication strategy buys the stock and borrows cash, adjusting the amount of stock and cash to replicate the cash flows on the option.

In practice, practitioners and researchers often confuse replication with hedging, generating a sign error.

### Realized variance and the assumptions of the Black-Scholes model

Dynamic hedgeing strategies only work if certain assumptions are satisfied. By contrast, static hedging strategies often work for a much broader set of assumptions.

What assumptions does the Black-Scholes model make?

The main assumption of the Black-Scholes model is that the underlying asset's price follows a geometric Brownian motion. There are two key ideas here:

1. **Volatility** ($\sigma$) is constant: it does not vary over time.

2. The asset's price path has **no jumps**: sample paths are continuous with probability one.

The geometric Brownian motion assumption can be relaxed somewhat. The Black-Scholes methodology:

1. Requires only minor adjustments if the volatility is a nonstochastic function of time; the approach does not work if volatility changes stochastically over time.

2. Does not require the risk premium or stock beta to be constant; both might change stochastically over time.

3. Is easily modified to handle a nonflat term structure of interest rates. 

4. Can be adapted to allow interest rates to be stochastic if interest rate changes are uncorrelated with changes in the asset's price.

The dynamic trading strategy works because the **realized variance** of a price path is a known constant. To see this, let $t_0=t$, $t_1 = t_0 + \Delta t$, $t_2 = t_0 + 2 \sm \Delta t$, ..., $t_N = t_0 + N \sm \Delta t = T$ partition the time span of the option hedging strategy into intervals of length $\Delta t = (T-t)/N$. Consider a dynamic hedging strategy which adjusts the hedge at the beginning of each time interval of length $\Delta t$. Define $\Delta P_{n+1} := P(t_{n+1}) - P(t_n)$. Intuitively, the Black-Scholes model works because the **realized variance** of the underlying, defined by $\sum_{n=1}^N \left( \frac{\Delta P_n}{P_n} \right)^2$ is a random variable which, in the limit $\Delta t \rightarrow 0$, has a known mean $\sigma^2 \sm (T - t)$ and a variance equal to zero.  



### Implied volatility

If we know the price of the option $c$ and the Black-Scholes inputs $P$, $K$, $m$, and $r_f$, but we do not know the volatility $\sigma$, we can obtain $\sigma$ as an **implied volatility** by solving the equation

$$
c = C(P, K, m, r_f, \sigma)
$$

for $\sigma$.  This works because $\sigma$ is the only unknown in this equation.

### Early exercise

**European options** must be exercised at maturity.  **American options** may be exercised early.  Obviously, American options are worth at least as much as otherwise equivalent European options. But are they worth more?

Here is some intuition for helping think about this problem:

1. If the rate of interest is positive, this favors delaying exercise of a call option but favors exercising a put option early.

2. If the stock pays dividends but the dividends are lost if the option is not exercised before the dividend is paid, this favors exercising call options earlybut delaying the exercise of put options.

3. The convexity of the option, by itself, implies that it is valuable to wait and see what happens and therefore not exercise early.

Conclusion: If the stock does not pay dividends and the rate of interest is positive, it is not optimal to exercise a call option early.  In this case, the value of a European call option is the same as the value of an American call option.

### Expected return on options

The Black-Scholes model gives the discounted expected payoff on the option under the hypothetical assumption that the risk premium on the stock is zero.

If $\beta$ and $\pi$ are known, then the expected payoff on the option with the actuarily correct
probabilities are given by changing the price in the Black-Scholes formula from $P$ to $P \sm \textrm{e}^{\beta \sm \pi \sm m}$.

To obtain the expected return on the option, solve the equation

$$
C(P, K, m, r_f, \sigma) = \textrm{e}^{-\beta_c \sm \pi \sm m} 
\sm C(P \sm \textrm{e}^{\beta \sm \pi \sm m}, K, m, r_f, \sigma)
$$

for $\beta_c$. The expected return is then $r_f + \beta_c \sm \pi$.

While this approach implements the misguided idea of discounting the expected payoff at a risk-adjusted discount rate based on $\beta_c$, it is easy to see that finding the correct option beta $\beta_c$ requires solving the problem using some other method, such as the Black-Scholes model.

### "Greeks"

Researchers and practitioners use Greek letters (including the nonexistent letter "vega" for "nu") to describe the derivatives of the option price with respect to various arguments.

The derivative of an option with respect to its price, $\frac{\partial C}{\partial P}$, is called the option's **delta**. The delta gives the hedge ratio used in a dynamic strategy to replicate the option. The delta multiplied by minus-one gives the short position in a dynamic hedging strategy.

It can the shown that the delta is the same as $\text{N}(d_1)$ from the Black-Scholes formula:

$$
\frac{\partial C}{\partial P} = \text{N}(d_1),
\qquad \text{where} \qquad
d_1 := \frac{\log \left( \frac{P}{K \sm \mathrm{e}^{-r_f \sm m}} \right) + \sigma^2 \sm m / 2}{\sigma \sm \sqrt{m}}
$$

Under the assumption of risk neutrality, the probability of option exercise is $\text{N}(d_2)$.

In the Black-Scholes formula, the term $P \sm \mathrm{N}(d_1)$ represents the price of an **asset-or-nothing call**, which is defined as a security which gives the owner of the asset one share of stock worth $P$ if the call option is exercised and zero otherwise. The term $K \sm \mathrm{e}^{-r_f \sm m} \sm \mathrm{N}(d_2)$ represents the price of a **cash-or-nothing call**, which is defined as a security which pays off the strike price $K$ if the call option is exercised, zero otherwise. The asset-or-nothing call and cash-or-nothing call are examples of **binary options**, which are **pure securities** which pay off a fixed amount or zero. Note that receiving one share of stock worth $P$ is a constant amount if the stock is used as the numeraire rather than cash.

$$
\begin{aligned}
C(P, K, m, r_f, \sigma) &= P \sm \mathrm{N}(d_1) - K \sm \mathrm{e}^{-r_f \sm m} \sm \mathrm{N}(d_2), \\
\text{where} \qquad d_1 &:= \frac{\log \left( \frac{P}{K \sm \mathrm{e}^{-r_f \sm m}} \right) + \sigma^2 \sm m / 2}{\sigma \sm \sqrt{m}},
\qquad
d_2 := d_1 - \sigma \sm\sqrt{m}.
\end{aligned}
$$
